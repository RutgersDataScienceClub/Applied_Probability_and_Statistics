# Monte Carlo - Estimating π

We estimate π by sampling random points in the unit square and counting the fraction that lie inside the quarter-circle of radius 1. By the law of large numbers the estimate improves as sample size increases.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Single-N experiment
N = 1000
np.random.seed(0)
x = np.random.rand(N)
y = np.random.rand(N)
inside = (x**2 + y**2) <= 1.0
pi_hat = 4 * inside.mean()
print(f'Pi estimate (N={N}): {pi_hat:.6f}, error: {abs(pi_hat - np.pi):.6f}')

# Visualization
plt.figure(figsize=(4,4))
plt.scatter(x[inside], y[inside], s=5, color='C0', label='inside')
plt.scatter(x[~inside], y[~inside], s=5, color='C1', label='outside')
theta = np.linspace(0, np.pi/2, 100)
plt.plot(np.cos(theta), np.sin(theta), color='k')
plt.axis('equal')
plt.legend()
plt.title('Monte Carlo points (quarter circle)')
plt.show()
```

```python
# Convergence experiment
N_values = [100, 300, 1000, 3000, 10000]
estimates = []
np.random.seed(1)
for N in N_values:
    x = np.random.rand(N)
    y = np.random.rand(N)
    est = 4 * ((x**2 + y**2) <= 1.0).mean()
    estimates.append(est)
    print(f'N={N:5d} pi_hat={est:.6f} error={abs(est - np.pi):.6f}')

plt.plot(N_values, estimates, marker='o')
plt.axhline(np.pi, color='k', linestyle='--', label='pi')
plt.xscale('log')
plt.xlabel('N (log scale)')
plt.ylabel('pi estimate')
plt.title('Convergence of Monte Carlo estimate')
plt.legend()
plt.show()
```

**Interpretation:** Monte Carlo estimation converges slowly (roughly proportional to 1/sqrt(N) in standard error). Larger `N` reduces the error but convergence is noisy for small sample sizes.
