# MLE - Estimating Normal Distribution Parameters

Maximum likelihood estimation (MLE) chooses parameters that maximize the likelihood of observed data. For a normal distribution, the MLE for the mean is the sample mean and for the standard deviation is the square root of the average squared deviation (no Bessel correction).

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_pdf(x, mu, sigma):
    return (1/(np.sqrt(2*np.pi)*sigma)) * np.exp(-0.5*((x-mu)/sigma)**2)
```

```python
# Generate synthetic data
mu_true = 2.0
sigma_true = 1.5
np.random.seed(0)
n = 50
data = np.random.normal(mu_true, sigma_true, size=n)

# MLE estimates
mu_hat = data.mean()
sigma_hat = np.sqrt(((data - mu_hat)**2).mean())
print(f'True mu={mu_true:.3f}, sigma={sigma_true:.3f}')
print(f'Estimated mu_hat={mu_hat:.3f}, sigma_hat={sigma_hat:.3f}')
```

```python
# Visualize histogram and PDFs
xs = np.linspace(data.min()-1, data.max()+1, 200)
plt.hist(data, bins=15, density=True, alpha=0.6, color='C0')
plt.plot(xs, normal_pdf(xs, mu_true, sigma_true), label='True PDF', color='C1')
plt.plot(xs, normal_pdf(xs, mu_hat, sigma_hat), label='Estimated PDF', color='C2', linestyle='--')
plt.legend()
plt.title('Data histogram with true and estimated PDFs')
plt.show()
```

```python
# Experiment: different sample sizes
for n in [10, 30, 100, 300]:
    np.random.seed(0)
    d = np.random.normal(mu_true, sigma_true, size=n)
    mu_h = d.mean()
    sigma_h = np.sqrt(((d - mu_h)**2).mean())
    print(f'n={n:4d} | |mu_err|={(abs(mu_h-mu_true)):.4f} | |sigma_err|={(abs(sigma_h-sigma_true)):.4f}')
```

**Interpretation:** As sample size grows, the MLE estimates for `mu` and `sigma` converge toward the true values. Small samples show greater variability.
