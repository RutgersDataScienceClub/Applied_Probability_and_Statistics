# Markov Chain - Simple Weather Model

A Markov chain's next state depends only on the current state. We'll simulate a 2-state weather model (`Sunny`, `Rainy`) using a transition matrix and compare empirical frequencies to the stationary distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

states = ['Sunny','Rainy']
P = np.array([[0.8, 0.2], [0.4, 0.6]])  # rows: current, cols: next
print('Transition matrix P:')
print(P)
```

```python
# Simulate trajectory
np.random.seed(0)
T = 30
cur = 0  # start Sunny
traj_idx = [cur]
for t in range(1, T):
    cur = np.random.choice([0,1], p=P[cur])
    traj_idx.append(cur)
traj = [states[i] for i in traj_idx]
print(' -> '.join(traj))

# Empirical distribution
counts = np.bincount(traj_idx, minlength=2)
empirical = counts / counts.sum()
print('Empirical distribution:', dict(zip(states, empirical)))
```

```python
# Stationary distribution: solve pi = pi P (left eigenvector)
w, v = np.linalg.eig(P.T)
idx = np.argmin(np.abs(w - 1.0))
stat = np.real(v[:, idx])
stat = stat / stat.sum()
print('Stationary distribution:', dict(zip(states, stat)))

# Bar chart comparison
x = np.arange(len(states))
width = 0.35
plt.bar(x - width/2, empirical, width, label='Empirical')
plt.bar(x + width/2, stat, width, label='Stationary')
plt.xticks(x, states)
plt.legend()
plt.title('Empirical vs Stationary distribution')
plt.show()
```

**Interpretation:** Over many steps, the empirical frequencies approach the stationary distribution. The stationary vector solves `pi = pi P` and is independent of initial state for an ergodic chain.
